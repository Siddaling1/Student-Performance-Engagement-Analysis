```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# **What is the average attendance % of students by class?**

```python
df.head()
```

```python
class_by_Avg =df.groupby('Class')['Attendance%'].mean()
class_by_Avg
```

```python
from collections.abc import Container
class_by_Avg=df.groupby('Class')['Attendance%'].mean().plot(x='Class',y='Attendance%',kind='line',title='Avg Attendance by Class')
```

# **Which subject has the highest average score?**

```python
df.head()
```

```python
Sub_Avg=df[['Math_Score','Science_Score','English_Score']].mean()
print(Sub_Avg)
```

```python
from ast import Sub
ax = Sub_Avg.plot.bar(color='orange')
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

# **What is the gender distribution across classes?**

```python
gen=df.groupby(['Class','Gender']).size().unstack(fill_value=0)
print(gen)
gen.plot(kind="bar")
```

# **Find the top 10 students based on total scores.**

```python
df["Total_Score"]=df['Math_Score']+df['Science_Score']+df['English_Score']
df.head()
```

```python
stu=df.groupby('StudentID')['Total_Score'].max().sort_values(ascending=False).head(10)
stu
```

# **Compare average study hours vs average final grade.**

```python
df.head()
```

```python
df['Grade_NO']=df['Final_Grade'].replace({"A":5,"B":4,"C":3,"D":2})
df.head()
```

```python
com=df.groupby('Final_Grade')['Study_Hours'].mean().plot(kind='bar',color='red')
com
```

# **Does higher attendance correlate with higher scores?**

```python
df.head()
```

```python
df['Attendance%'].max()
```

```python
df['Attendance%'].min()
```

```python
bins=[50,60,70,80,90,100]
labels=['50-59','60-69','70-79','80-89','90-100']
df['Attendance_Group']=pd.cut(df['Attendance%'],bins=bins,labels=labels,right=True, include_lowest=True)
df.head()
```

```python
Att=df.groupby('Attendance_Group')['Total_Score'].mean().plot(kind='bar')
```

# **Is there a significant difference in scores between students with and without extracurricular activities?**

# **Does parental education level influence student performance?**

```python
m=df.groupby('Parent_Education')['Total_Score'].mean().plot(kind="bar",color="lightblue")
for container in m.containers:
  m.bar_label(container)
```

# **Build age groups and compare their performance.**

```python
df.head()
```

```python
df['Age'].max()
```

```python
df['Age'].min()
```

```python
bins=[13,15,17,19]
labels=['13-14','15-16','17-18']
df['Age_Group']=pd.cut(df['Age'],bins=bins,labels=labels,right=True,include_lowest=True)
df.head()
```

```python
df.groupby('Age_Group')['Total_Score'].sum().plot(kind='barh',color='yellow')
```
